```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image using cv2
img = cv2.imread('images.jpeg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Get image dimensions
h, w = img_gray.shape
```

```python
# identity transformation
iden_img = cv2.flip(img, 1)
```

```python
# reflect trandformation
ref_img = cv2.flip(img, 0)
```

```python
# scaling transformation
s_x, s_y = 0.75, 1.25
img_scaled = cv2.resize(img, None, fx=s_x, fy=s_y, interpolation = cv2.INTER_LINEAR)
```

```python
# rotate transformation
theta = np.pi / 6
mat_rotate = cv2.getRotationMatrix2D((w/2, h/2), np.degrees(theta), 1)
rot_image = cv2.warpAffine(img_scaled, mat_rotate, (w, h))
```

```python
# shear transformation
M_shear = np.float32([[1, 0.5, 0], [0.5, 1, 0]])
she_img = cv2.warpAffine(rot_image, M_shear, (w, h))
```

```python
# Display the images using matplotlib
plt.figure(figsize=(12, 8), facecolor = 'gray')

plt.subplot(2, 3, 1)
plt.title('Original')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.title('Reflection')
plt.imshow(iden_img, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.title('Rotation')
plt.imshow(ref_img, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.title('Scaling')
plt.imshow(img_scaled, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.title('Affine')
plt.imshow(rot_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.title('Shearing')
plt.imshow(she_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
