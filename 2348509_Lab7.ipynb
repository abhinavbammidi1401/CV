`Reading the image`

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# read the image
original = cv2.imread('blessed.webp',0)

#plt.subplot(2,1,1)
plt.title('Original Image')
plt.imshow(original, cmap='gray')
plt.axis("off")
```

`Salt pepper noise`

```python
# get the image size
img_size = original.size
# set the percentage of pixels that should contain noise
noise_percentage = 0.1
# determine the size of the noise based on the noise percentage
noise_size = int(noise_percentage * img_size)
# randomly selelct indices for adding noise
random_indices = np.random.choice(img_size, noise_size)
# create a copy of the original image that serves
# as a template for the noised image
noisy = original.copy()
# create a noise list with random placements of min and max values
# of the image
noise_sp = np.random.choice([original.min(), original.max()], noise_size)
# replace the values of the templated noise image at 
# random indices with the noise to obtain the final noisy image
noisy.flat[random_indices] = noise_sp

plt.figure(figsize=(10,8))
plt.style.use("dark_background")

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(original, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title('Salt-Pepper')
plt.imshow(noisy, cmap='gray')
plt.axis("off")
```

`Random Noise`

```python
noise_gs = np.random.normal(0, 100, original.shape)

img_noised = original + noise_gs
img_noised = np.clip(img_noised, 0, 255).astype(np.uint8)

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(original, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Random Noise")
plt.imshow(img_noised, cmap='gray')
plt.axis("off")
```

`Applying Median Blur`

```python
# apply median blur
denoised = cv2.medianBlur(img_noised, 5)
denoised_sp = cv2.medianBlur(noisy, 5)

plt.figure(figsize=(20,8))

plt.subplot(2, 5, 1)
plt.title("Original Image")
plt.imshow(original, cmap='gray')
plt.axis("off")

plt.subplot(2, 5, 2)
plt.title("Random Noise")
plt.imshow(img_noised, cmap='gray')
plt.axis("off")

plt.subplot(2, 5, 3)
plt.title("Salt-Pepper Noise")
plt.imshow(noisy, cmap='gray')
plt.axis("off")

plt.subplot(2, 5, 4)
plt.title("Median Blur on Salt-pepper")
plt.imshow(denoised, cmap='gray')
plt.axis("off")

plt.subplot(2, 5, 5)
plt.title("Median Blur on Random Noise")
plt.imshow(denoised_sp, cmap='gray')
plt.axis("off")
```

`Applying Gaussian blur`

```python
# apply median blur
denoised_g = cv2.GaussianBlur(img_noised, (5,5), 0)
denoised_sp_g = cv2.GaussianBlur(noisy, (5,5), 0)

plt.figure(figsize=(20,8))

plt.subplot(2, 5, 1)
plt.title("Original Image")
plt.imshow(original, cmap='gray')
plt.axis("off")

plt.subplot(2, 5, 2)
plt.title("Random Noise")
plt.imshow(img_noised, cmap='gray')
plt.axis("off")

plt.subplot(2, 5, 3)
plt.title("Salt-Pepper Noise")
plt.imshow(noisy, cmap='gray')
plt.axis("off")

plt.subplot(2, 5, 4)
plt.title("Gaussian Blur on Salt-pepper")
plt.imshow(denoised_g, cmap='gray')
plt.axis("off")

plt.subplot(2, 5, 5)
plt.title("Gaussian Blur on Random Noise")
plt.imshow(denoised_sp_g, cmap='gray')
plt.axis("off")
```
