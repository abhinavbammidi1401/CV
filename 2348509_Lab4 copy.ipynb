**Average Filter**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def avg_filter(image, kernel_size):
    return cv2.blur(image, (kernel_size, kernel_size))

image = cv2.imread('salterpepper.jpg')

filtered_image = avg_filter(image, kernel_size=3)

# Display all the output images
plt.figure(figsize=(12, 12))

plt.subplot(4, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(4, 3, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title('Average Filter - In Built')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2 
import numpy as np 
   
      
# Read the image 
img = cv2.imread('salterpepper.jpg', 0) 
  
# Obtain number of rows and columns  
# of the image 
m, n = img.shape 
   
# Develop Averaging filter(3, 3) mask 
mask = np.ones([3, 3], dtype = int) 
mask = mask / 9
   
# Convolve the 3X3 mask over the image  
img_new = np.zeros([m, n]) 
  
for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2] 
         
        img_new[i, j]= temp 
          
img_new = img_new.astype(np.uint8) 

# Display all the output images
plt.figure(figsize=(12, 12))

plt.subplot(4, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(4, 3, 2)
plt.imshow(img_new, cmap='gray')
plt.title('Average Filter - Manual')
plt.axis('off')

plt.tight_layout()
plt.show()
```

**Gaussian Blur**

```python
import cv2
import numpy as np

image = cv2.imread("salterpepper.jpg", 0)

Gaussian = cv2.GaussianBlur(image, (7, 7), 0)

# Display all the output images
plt.figure(figsize=(12, 12))

plt.subplot(4, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(4, 3, 2)
plt.imshow(Gaussian, cmap='gray')
plt.title('Gaussian Blur - In Built')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

def manual_gaussian(image, sigma):
    kernel_size = 2 * int(3 * sigma) + 1
    x = np.arange(-kernel_size // 2 + 1, kernel_size // 2 + 1, dtype=float)
    y = np.arange(-kernel_size // 2 + 1, kernel_size // 2 + 1, dtype=float)
    x2 = x**2
    y2 = y**2
    gaussian = np.exp(-(x2 + y2) / (2 * sigma**2)) / (2 * np.pi * sigma**2)
    kernel = np.outer(gaussian, gaussian)
    
    if len(image.shape) == 2:
        filtered_image = cv2.filter2D(image, -1, kernel)
    else:
        filtered_image = np.zeros_like(image)
        for channel in range(image.shape[2]):
            filtered_image[:, :, channel] = cv2.filter2D(image[:, :, channel], -1, kernel)
    return filtered_image

image = cv2.imread("images.jpeg", 0)
sigma = 2

manual_gauss = manual_gaussian(image, sigma)

# Display all the output images
plt.figure(figsize=(12, 12))

plt.subplot(4, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(4, 3, 2)
plt.imshow(manual_gauss, cmap='gray')
plt.title('Gaussian Blur - Manual')
plt.axis('off')

plt.tight_layout()
plt.show()  
```

**Laplacian Filter**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('moon.jpg', 0)

kernel = np.array([[0, 1, 0],
                   [1, -4, 1],
                   [0, 1, 0]])

LaplacianImage = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)

# Display all the output images
plt.figure(figsize=(12, 12))

plt.subplot(4, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(4, 3, 2)
plt.imshow(LaplacianImage, cmap='gray')
plt.title('Laplacian Filter - Manual')
plt.axis('off')

plt.tight_layout()
plt.show()
```
