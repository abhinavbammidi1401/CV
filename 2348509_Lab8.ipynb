```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("images.jpeg", 0)
```

```python
# apply sobel edge detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
sobel_x = cv2.normalize(sobel_x, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_64F)
sobel_y = cv2.normalize(sobel_y, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_64F)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)
```

```python
# apply Prewitt Edge detection
prewitt_kernel_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_kernel_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
prewitt_x = cv2.filter2D(image, -1, prewitt_kernel_x)
prewitt_y = cv2.filter2D(image, -1, prewitt_kernel_y)
prewitt_x = cv2.normalize(prewitt_x, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_64F)
prewitt_y = cv2.normalize(prewitt_y, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_64F)
prewitt_combined = cv2.magnitude(prewitt_x, prewitt_y)
```

```python
# apply Roberts edge detection
roberts_kernel_x = np.array([[1, 0], [0, -1]])
roberts_kernel_y = np.array([[0, 1], [-1, 0]])
roberts_x = cv2.filter2D(image, -1, roberts_kernel_x)
roberts_y = cv2.filter2D(image, -1, roberts_kernel_y)
roberts_x = cv2.normalize(roberts_x, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_64F)
roberts_y = cv2.normalize(roberts_y, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_64F)
roberts_combined = cv2.magnitude(roberts_x, roberts_y)
```

```python
# apply laplacian edge detection
laplacian = cv2.Laplacian(image, cv2.CV_64F)
```

```python
# apply Canny Edge Detection
canny = cv2.Canny(image, 100, 200)
```

```python
# Plot the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Edge Detection')

plt.subplot(2, 3, 2)
plt.imshow(prewitt_combined, cmap='gray')
plt.title('Prewitt Edge Detection')

plt.subplot(2, 3, 3)
plt.imshow(roberts_combined, cmap='gray')
plt.title('Roberts Edge Detection')

plt.subplot(2, 3, 4)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian Edge Detection')

plt.subplot(2, 3, 5)
plt.imshow(canny, cmap='gray')
plt.title('Canny Edge Detection')

plt.show()
```
